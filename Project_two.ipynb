```python
# Note: We use only odd numbers because we don't want any ties.
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_resampled_train_scaled, y_resampled_train)
    train_score = knn.score(X_resampled_train_scaled, y_resampled_train)
    test_score= knn.score(X_resampled_test_scaled, y_resampled_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")

# Plot the results
plt.plot(range(1, 20, 2), train_scores, marker='o', label="training scores")
plt.plot(range(1, 20, 2), test_scores, marker="x", label="testing scores")
plt.xlabel("k neighbors")
plt.ylabel("accuracy score")
plt.legend()
plt.show()
```

```python
def process_data(df):
    X = df.drop(['Diabetes_binary'], axis=1)
    y = df['Diabetes_binary']
    X_train, X_test, y_train, y_test = train_test_split(X, y)
    return X_train, X_test, y_train, y_test
```

```python
def over_sampling (X, y):
  random_oversampler = RandomOverSampler()
  X_over, y_over = random_oversampler.fit_resample(X, y)
  return X_over, y_over
```

```python
def under_sampling (X, y):
  random_undersampler = RandomUnderSampler()
  X_under, y_under = random_undersampler.fit_resample(X, y)
  return X_under, y_under
```

```python
def pipe (X, y, model):
    X_train, X_test, y_train, y_test = process_data(df)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    train_acc_score=accuracy_score(y_train,y_pred_train)
    test_acc_score=accuracy_score(y_test,y_pred)
    print("Accuracy Score:")
    print(accuracy_score(y_test, y_pred))
    print(f"Test Accuracy Score: {test_acc_score}")
    print(f"Train Accuracy Score:{train_acc_score}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Balanced Acc. Score:")
    print(balanced_accuracy_score(y_test, y_pred))
```
